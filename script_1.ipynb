```python
import pandas as pd

df = pd.read_excel('data.xlsx')
df.head()

df = df[df['Year']!=1900]
```

# 2022 Expenses per Sub-Category 

```python
import seaborn as sns
import matplotlib.pyplot as plt

plot_1 = df[df['Year']==2022].groupby(['Sub-Category']).sum()

fig, ax = plt.subplots(figsize=(30, 5))
sns.barplot(data=plot_1, x=plot_1.index, y="Debits")
plt.show()
```

# 2022 Expenses per Category

```python
import seaborn as sns
import matplotlib.pyplot as plt

plot_2 = df[df['Year']==2022].groupby(['Category']).sum()

fig, ax = plt.subplots(figsize=(13, 5))
sns.barplot(data=plot_2, x=plot_2.index, y="Debits")
plt.show()
```

# 2022 6 categories per month comparison

```python
import numpy as np
import matplotlib.pyplot as plt
 
def get_values(df,category):
    a=1
    values=[]
    for i in range(12):
        data=df[df['Month']==a]
        values.append(data[data['Category']==str(category)]['Debits'].sum())
        a=a+1
    return values    
    
N = 12
ind = np.arange(N) 
width = 0.13
plt.figure(figsize=(20,8))


dining=get_values(df,str('Dining Out'))
bar1 = plt.bar(ind, dining, width, color = 'b')

discret=get_values(df,str('Discretionary'))
bar2 = plt.bar(ind+width, discret, width, color = 'r')

#income = get_values(plot_3,2)
#bar3 = plt.bar(ind+width*2, income, width, color = 'b')

living_ex=get_values(df,str('Living Expenses'))
bar4 = plt.bar(ind+width*2, living_ex, width, color = 'c')

transport=get_values(df,str('Transport'))
bar5 = plt.bar(ind+width*3, transport, width, color = 'm')

medical=get_values(df,str('Medical'))
bar6 = plt.bar(ind+width*4, medical, width, color = 'y')

savings=get_values(df,str('Investments/Savings'))
bar7 = plt.bar(ind+width*5, savings, width, color = 'k')

  
plt.xlabel("Month")
plt.ylabel('Amount')
plt.title("Expenses per category each month")
  
plt.xticks(ind+width,['January','February','March','April','May','june','July','August','September','October','November','December'])
plt.legend( (bar1, bar2,bar4,bar5,bar6,bar7), ('Dining Out', 'Discrecionary','Living Expenses','Transport','Medical','Savings') )
plt.grid()
plt.show()
```
